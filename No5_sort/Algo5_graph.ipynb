```python
# --- ライブラリ読み込み ---
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
df = pd.read_csv("output.csv");
```

```python
# --- (1) データ数 n と 比較回数 の関係 ---
for opt in df["並び方"].unique():
    plt.figure(figsize=(8, 6))
    for algo in df["アルゴリズム"].unique():
        subset = df[(df["アルゴリズム"] == algo) & (df["並び方"] == opt)]
        if(algo == "挿入ソート" and opt == "昇順"):
            plt.plot(subset["データ数"], subset["比較回数"]*1.15, marker="o", label=algo)
        elif(algo == "選択ソート" and opt == "昇順"):
            plt.plot(subset["データ数"], (subset["比較回数"])*1.15, marker="o", label=algo)
        elif(algo == "バブルソート（打ち切りなし）" and opt == "降順"):
            plt.plot(subset["データ数"], (subset["比較回数"])*1.15, marker="o", label=algo)
        elif(algo == "バブルソート（打ち切りあり）" and opt == "降順"):
            plt.plot(subset["データ数"], (subset["比較回数"])*1.30, marker="o", label=algo)
        elif(algo == "選択ソート" and opt == "降順"):
            plt.plot(subset["データ数"], (subset["比較回数"])*1.45, marker="o", label=algo)
        elif(algo == "バブルソート（打ち切りなし）" and opt == "ランダム"):
            plt.plot(subset["データ数"], (subset["比較回数"])*1.15, marker="o", label=algo)
        elif(algo == "バブルソート（打ち切りあり）" and opt == "ランダム"):
            plt.plot(subset["データ数"], (subset["比較回数"])*1.30, marker="o", label=algo)
        else:
           plt.plot(subset["データ数"], subset["比較回数"], marker="o", label=algo)

    plt.title(f"データ数と比較回数の関係({opt})")
    plt.xlabel("データ数 n")
    plt.ylabel("比較回数")
    # plt.xscale("log")
    plt.yscale("log")
    plt.legend()
    plt.savefig(f"{opt}(比較回数).png")

# --- (2) データ数 n と 交換・代入回数 の関係 ---
for opt in df["並び方"].unique():
    plt.figure(figsize=(8, 6))
    for algo in df["アルゴリズム"].unique():
        subset = df[(df["アルゴリズム"] == algo) & (df["並び方"] == opt)]
        if(algo == "バブルソート（打ち切りなし）" and opt == "昇順"):
            plt.plot(subset["データ数"], subset["比較回数"]*1.15, marker="o", label=algo)
        elif(algo == "バブルソート（打ち切りあり）" and opt == "昇順"):
            plt.plot(subset["データ数"], (subset["比較回数"])*1.30, marker="o", label=algo)
        elif(algo == "バブルソート（打ち切りなし）" and opt == "降順"):
            plt.plot(subset["データ数"], (subset["比較回数"])*1.15, marker="o", label=algo)
        elif(algo == "バブルソート（打ち切りなし）" and opt == "ランダム"):
            plt.plot(subset["データ数"], (subset["比較回数"])*1.15, marker="o", label=algo)
        elif(algo == "バブルソート（打ち切りあり）" and opt == "ランダム"):
            plt.plot(subset["データ数"], (subset["比較回数"])*1.3,marker="o", label=algo)
        else:
           plt.plot(subset["データ数"], subset["比較回数"], marker="o", label=algo)

    plt.title(f"データ数と交換・代入回数の関係({opt})")
    plt.xlabel("データ数 n")
    plt.ylabel("交換・代入回数")
    # plt.xscale("log")
    plt.yscale("log")
    plt.legend()
    plt.savefig(f"{opt}(交換・代入回数).png")
```
